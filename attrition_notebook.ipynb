# Visualization

### Overall Attrition

```python
# Count employees by attrition
attrition_count = df["Attrition"].value_counts()
attrition_count
```

```python
# Create the Pie Chart
plt.figure(figsize=(7, 7))

plt.pie(attrition_count, labels = attrition_count.index, autopct="%1.1f%%",startangle=90,colors=['#2ecc71', '#e74c3c'])

plt.title("Overall Employee Attrition")

plt.savefig("overall_employee_attrition.png", dpi=300, bbox_inches="tight")

plt.show()
```

##### Insight
  
##### - The company has 1,470 employees.
##### - 237 employees (16.1%) left the company, while 1,233 employees (83.9%) stayed.
##### - Most employees continue working in the company.
##### - Only a small percentage of employees leave.
##### - Finding the reasons behind these resignations can help improve employee retention.

### Employee Attrition by Age Group

```python
# create age groups
df["Age_Group"] = pd.cut(
    df["Age"], 
    bins = [18, 24, 34, 44, 54, 64], 
    labels = ['Under 25', '25 - 34', '35 - 44', '45 - 54', '55+'], 
    include_lowest = True)

attrition_by_age = df.groupby('Age_Group')["Attrition"].value_counts().unstack()
attrition_by_age
```

```python
# create a grouped (joined) bar chart:
x = np.arange(len(attrition_by_age.index))
width = 0.35

plt.figure(figsize=(8,5))

bars1 = plt.bar(x - width/2, attrition_by_age["No"], width, label="Stayed",  color="#2ecc71")
bars2 = plt.bar(x + width/2, attrition_by_age["Yes"], width, label="Left", color="#e74c3c")

plt.title("Employee Attrition by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Number of Employees")
plt.xticks(x, attrition_by_age.index)
plt.legend()

# Add value labels
for bar in bars1:
    plt.text(
        bar.get_x() + bar.get_width()/2,
        bar.get_height(),
        int(bar.get_height()),
        ha="center",
        va="bottom"
    )

for bar in bars2:
    plt.text(
        bar.get_x() + bar.get_width()/2,
        bar.get_height(),
        int(bar.get_height()),
        ha="center",
        va="bottom"
    )

plt.savefig("employee_attrition_by_age_group.png", dpi=300, bbox_inches="tight")

plt.show()
```

##### Insight:

##### - Employees aged 25–34 have the highest number of resignations.
##### - Employees aged 55+ have the lowest number of resignations.
##### - Younger employees leave the company more often.
##### - Older employees are more likely to stay with the company.
##### - The company should focus on retaining younger employees.

### Attrition Rate by Department

```python
# create an Attrition Flag
df["Attrition_Flag"] = df["Attrition"].replace({'Yes': 1, 'No': 0})
department_attrition = (df.groupby('Department').agg(Total_Attrition = ("Attrition_Flag", 'sum'), Total_Employee = ("EmployeeNumber", 'count')))

# calculate Attrition Rate
department_attrition["AttritionRate"] = (department_attrition["Total_Attrition"] / department_attrition["Total_Employee"] * 100).round(2)

department_attrition = department_attrition.reset_index()
department_attrition
```

```python
# Plot the Bar Chart
plt.figure(figsize=(8,5))

bars =plt.bar(department_attrition['Department'], department_attrition["AttritionRate"], color=['#e74c3c', '#3498db', '#f39c12'])

plt.title("Attrition Rate by Department")

plt.xlabel("Department")
plt.ylabel("Attrition Rate (%)")

# value labels
for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height(),
        str(bar.get_height()) + "%",
        ha="center",
        va="bottom"
    )

plt.savefig("attrition_rate_by_department.png", dpi=300, bbox_inches="tight")


plt.show()
```

##### Insight:

##### - The Sales department has the highest attrition rate (20.63%).
##### - The Research & Development department has the lowest attrition rate (13.84%).
##### - Employee turnover is different across departments.
##### - Sales employees leave more often than employees in other departments.
##### - The company should improve employee retention in the Sales department.

### Attrition by Monthly Income 

```python
# create a box plot
plt.figure(figsize=(8,5))

box = plt.boxplot([df[df["Attrition"] == "No"]["MonthlyIncome"],df[df["Attrition"] == "Yes"]["MonthlyIncome"]],tick_labels=["Stayed", "Left"], patch_artist=True)

box["boxes"][0].set_facecolor("#2ecc71")
box["boxes"][1].set_facecolor("#e74c3c")

plt.title("Monthly Income vs Attrition")
plt.xlabel("Attrition")
plt.ylabel("Monthly Income")

plt.savefig("monthly_income_vs_attrition.png", dpi=300, bbox_inches="tight")


plt.show()
```

##### Insight:

##### - Employees who left the company usually had lower salaries.
##### - Employees who stayed generally earned higher salaries.
##### - Lower salary may be one reason employees leave.
##### - A few high-income employees also left, but they are very few.
##### - Better salary and career growth may help reduce attrition.

### Attrition by OverTime 

```python
# create the summary table
attrition_overtime= df.groupby('OverTime')['Attrition'].value_counts().unstack()
attrition_overtime
```

```python
# plot a grouped bar chart
x = range(len(attrition_overtime.index))
width = 0.35

plt.figure(figsize=(8,5))

bars1 = plt.bar([i - width/2 for i in x],attrition_overtime["No"],width,label="Stayed", color="#2ecc71")

bars2 = plt.bar([i + width/2 for i in x],attrition_overtime["Yes"],width,label="Left", color="#e74c3c")

plt.title("OverTime vs Attrition")
plt.xlabel("OverTime")
plt.ylabel("Number of Employees")

plt.xticks(x, attrition_overtime.index)
plt.legend()

# Value labels
for bars in [bars1, bars2]:
    for bar in bars:
        plt.text(
            bar.get_x() + bar.get_width()/2,
            bar.get_height(),
            int(bar.get_height()),
            ha="center",
            va="bottom"
        )

plt.savefig("overtime_vs_attrition.png", dpi=300, bbox_inches="tight")


plt.show()
```

##### Insight:

##### - Employees who work overtime leave the company more often.
##### - Employees without overtime are more likely to stay.
##### - Working extra hours may increase stress.
##### - Too much overtime can affect employee satisfaction.
##### - Reducing overtime may help lower attrition.

### Attrition Rate by Years at Company

```python
# create a flag
df["Attrition_Flag"] = df["Attrition"].replace({'Yes': 1, 'No': 0})

# Calculate attrition rate by years at company
years_attrition = (df.groupby("YearsAtCompany")["Attrition_Flag"].mean().mul(100).round(2).reset_index())

years_attrition
```

```python
# Plot the line chart
plt.figure(figsize=(8,5))

plt.plot(years_attrition["YearsAtCompany"],years_attrition["Attrition_Flag"],marker="o", color='#e74c3c', linewidth=2)

plt.title("Attrition Rate by Years at Company")
plt.xlabel("Years at Company")
plt.ylabel("Attrition Rate (%)")

plt.grid()

plt.savefig("attrition_rate_by_years_at_company.png", dpi=300, bbox_inches="tight")


plt.show()
```

##### Insight:

##### - Most employees leave during their first few years at the company.
##### - The attrition rate becomes lower as employees spend more years in the company.
##### - Employees with longer experience are more likely to stay.
##### - The first few years are important for employee retention.
##### - The company should support new employees during this period.

### Attrition by Job Satisfaction 

```python
# create the summary table
job_attrition = df.groupby("JobSatisfaction")["Attrition"].value_counts().unstack()
job_attrition
```

```python
# Plot the join bar graph
job_attrition.columns = ["Stayed", "Left"]
bars = job_attrition.plot(kind="bar",figsize=(8, 5), color=["#2ecc71", "#e74c3c"])

plt.title("Job Satisfaction vs Attrition")
plt.xlabel("Job Satisfaction")
plt.ylabel("Number of Employees")
plt.xticks(rotation=0)
plt.legend(title="Attrition")

# Value labels
for bars in bars.containers:
    for bar in bars:
        plt.text(
            bar.get_x() + bar.get_width()/2,
            bar.get_height(),
            int(bar.get_height()),
            ha="center",
            va="bottom"
        )

plt.savefig("job_satisfaction_vs_attrition.png", dpi=300, bbox_inches="tight")


plt.show()
```

##### Insight

##### - Employees with low job satisfaction leave more often.
##### - Employees with high job satisfaction are more likely to stay.
##### - Job satisfaction has an impact on employee retention.
##### - Improving employee satisfaction may reduce attrition.
##### - A positive work environment can help keep employees longer.

### Attrition by Job Role

```python
# reate the summary table
jobrole_attrition = (df.groupby("JobRole")["Attrition"].value_counts().unstack())

jobrole_attrition
```

```python
# Plot the join bar graph
jobrole_attrition.columns = ["Stayed", "Left"]

bars = jobrole_attrition.plot(kind="bar",figsize=(12,6), color=["#2ecc71", "#e74c3c"])

plt.title("Attrition by Job Role")
plt.xlabel("Job Role")
plt.ylabel("Number of Employees")
plt.xticks(rotation=45, ha="right")
plt.legend(title="Employee Status")

# Value labels
for bars in bars.containers:
    for bar in bars:
        plt.text(
            bar.get_x() + bar.get_width()/2,
            bar.get_height(),
            int(bar.get_height()),
            ha="center",
            va="bottom",
            fontsize=8
        )

plt.tight_layout()

plt.savefig("attrition_by_job_role.png", dpi=300, bbox_inches="tight")


plt.show()
```

##### Insight
##### - Employee attrition varies across different job roles.
##### - Some job roles have more employees leaving than others, while a few roles show very low attrition.
##### - This indicates that the nature of the job may influence employee turnover.
##### - The company should focus on the job roles with the highest attrition to understand the reasons behind employee resignations.
##### - Improving employee satisfaction and career growth opportunities in these roles may help reduce attrition.
